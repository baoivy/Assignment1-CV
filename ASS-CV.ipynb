# Import Libray 

```python
import numpy as np
from PIL import Image
```

# Load Image

```python
IMG_DIR = 'pic2.jpg'
```

```python
img = Image.open(IMG_DIR)
img
```

# Implement
Split RGB from picture, merge RGB to reconstruct the original image and create gray image from color image  

```python
def split_rgb(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    return r,g,b
```

```python
def merge_color(img_red, img_green, img_blue):
    new_img = np.stack([img_red, img_green, img_blue], axis=-1)
    return new_img
```

```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = r * 0.2989 + g * 0.5870 + b * 0.1140

    return gray
```

# Result

```python
img_array = np.array(img)
r_img, g_img, b_img = split_rgb(img_array)
```

```python
r_img = Image.fromarray(r_img.astype('uint8'))
r_img.save('r_img.jpg')

b_img = Image.fromarray(b_img.astype('uint8'))
b_img.save('b_img.jpg')

g_img = Image.fromarray(g_img.astype('uint8'))
g_img.save('g_img.jpg')
```
